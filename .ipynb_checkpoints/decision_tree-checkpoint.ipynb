```python
# decision boundaries if values scattered throughtout plot
# order of selection of attributes matter
# order selection by seeing entropy ; low entropy preferred (gives high information game)
# gini impurity ; Impurities within defined boundaries
```

```python
import pandas as pd
df=pd.read_csv("train.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop("Cabin",axis=1,inplace=True)
df
```

```python
df.Embarked.fillna(df.Embarked.mode()[0],inplace=True)
df.Age.fillna(int(df.Age.mean()),inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
l_e=LabelEncoder()
df["Sex_enc"]=l_e.fit_transform(df["Sex"])
```

```python
df["Fare"]=df["Fare"].astype("int64")
df.info()
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
df.drop(["PassengerId","Name","Sex","SibSp","Parch","Ticket","Embarked"],axis=1,inplace=True)
df
```

```python
model.fit(df[["Pclass","Sex_enc","Age","Fare"]],df["Survived"])
```

```python
model.predict([[1,0,38,71]])
```

```python
model.score(df.drop("Survived",axis=1),df.Survived)
```
